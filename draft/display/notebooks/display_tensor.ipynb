```python
def vis_gs(gs):
    """Visualization of a gs
    """
    x = 0
    size = len(gs.board)
    while x < size:
        y = 0
        string = ""
        while y < size:
            cell = gs.board[y, x]
            if cell == -1:
                string += "W"
            if cell == 0:
                string += "."
            if cell == 1: 
                string += "B"
            string += " "
            y += 1
        print(string) 
        x += 1
    if gs.current_player == 1:
        print("current_player = BLACK")
    if gs.current_player == -1:
        print("current_player = WHITE")
```

```python
def vis_layer(layer,num_layer):
    """Visualization of a layer
    """
    x = 0
    size = len(layer[num_layer, :, :])
    while x < size:
        string = ""
        y = 0
        while y < size:
            if layer[num_layer, y, x] == 1:
                string = string + " 1"    
            else:
                string = string + " ."
            y += 1
        print(string)
        x+=1
```

## Stone Colour

```python
colour = ft.stone_color_feature(gs)
print("Black")
vis_gs_layer(gs, colour, 0)
print("White")
vis_gs_layer(gs, colour, 1)
print("Empty")
vis_gs_layer(gs, colour, 2)
```

## Turns Since

```python
turns = ft.turns_since_move(gs)
i = 0
while i < 8:
    if i ==  7:
        print("Turns since move >= {}".format(i))
    else:
        print("Turns since move = {}".format(i))
    vis_gs_layer(gs, turns, i)
    i += 1
```
