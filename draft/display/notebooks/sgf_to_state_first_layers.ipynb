```python
def get_board(state, size):
    planes = np.zeros((3, size, size))
    x = 0
    while x < size:
        y = 0
        while y < size:
            cell = repr(state[x+1, y+1])    # conversion in string in order to be able to compare ==
            if cell == 'Black':
                planes[0, x, y] = 1 
            if cell == 'White':
                planes[1, x, y] = 1 
            if cell=='Empty':
                planes[2, x, y] = 1 
            y += 1
        x += 1
    return planes
```
